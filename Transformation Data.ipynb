```python
import pandas as pd
```

```python
data = pd.read_csv('world_population.csv')
data.head()
```

```python
data.tail()
```

```python
data.shape
```

```python
data.info()
```

```python
#checking the null values
data.isnull().sum()
```

```python
#replace null values with median

data['Median Age'].fillna(data['Median Age'].median(), inplace=True)
```

```python
data.isnull().sum()
```

```python
# convert population column datatype from object to numeric
data['Population'] = pd.to_numeric(data['Population'].str.replace(',',''))
```

```python
data.head()
```

```python
# convert Land Area column datatype from object to numeric
data['Land Area(Km2)'] = pd.to_numeric(data['Land Area (Km2)'].str.replace(',',''))
```

```python
data.head()
```

```python
# convert Density column datatype from object to numeric
data['Density (P/km2)'] = pd.to_numeric(data['Density (P/km2)'].str.replace(',',''))
```

```python
data.head()
```

```python
data.info()
```

```python
# save the dataset to a csv file
data.to_csv('world_population_2023.csv', index=False)
```
